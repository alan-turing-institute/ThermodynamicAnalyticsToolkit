# Optimizing the network - HMC test

```python

from ThermodynamicAnalyticsToolkit.models.model import model
import numpy as np

FLAGS = model.create_mock_flags(
    batch_data_files=["dataset-twoclusters.csv"],
    batch_size=20,
    max_steps=1000,
    output_activation="linear",
    inverse_temperature=1e1,
    hamiltonian_dynamics_steps=10,
    sampler="HamiltonianMonteCarlo",
    seed=426,
    step_width=1e-2
)
nn = model(FLAGS)
nn.init_network(None, setup="sample")
run_info, trajectory = nn.sample(return_run_info=True, return_trajectories=True)
nn.finish()


```

```python
display(trajectory)
```

```python
traj = np.array(trajectory.values)
print(traj.shape)

trajectory.to_csv("run_HMC.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

df_run = pd.read_csv("run_HMC.csv", sep=',', header=0)
run=np.asarray(df_run.loc[:,['step','loss','weight0', 'weight1', 'bias0']])
plt.plot(run[:,0], run[:,1])
plt.xlabel('steps')
plt.ylabel('loss')
plt.savefig('loss.png', bbox_inches='tight')
plt.show()

plt.plot(run[:,0], run[:,2])
plt.xlabel('steps')
plt.ylabel('weight 0')
plt.savefig('w0.png', bbox_inches='tight')
plt.show()

plt.plot(run[:,0], run[:,3])
plt.xlabel('steps')
plt.ylabel('weight 1')
plt.savefig('w1.png', bbox_inches='tight')
plt.show()

plt.plot(run[:,0], run[:,4])
plt.xlabel('steps')
plt.ylabel('bias 0')
plt.savefig('b0.png', bbox_inches='tight')
plt.show()
```
